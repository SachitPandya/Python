```python
df['Quater']=df['TransactionDate'].dt.quarter
Quater_Sales=df.groupby(["ProductID","Quater"]).agg(Quater_Sales_of_Product=("ProductSalesAmount","sum"))
Quater_Sales
```

```python
Max_Quaterly_Sales=Quater_Sales.where(Quater_Sales['Quater_Sales_of_Product']==Quater_Sales['Quater_Sales_of_Product'].max())
Max_Quaterly_Sales.dropna()
```
