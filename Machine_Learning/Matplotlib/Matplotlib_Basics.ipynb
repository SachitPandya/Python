- Bar Graph()

```python
df1=df.groupby("neighbourhood_group").agg(Average_reviews=("review_rate_number",'mean'))
n1=df['neighbourhood_group'].unique()
n1.sort()
df1,n1
```

```python
plt.bar(n1,df1['Average_reviews'])
plt.ylim(3,3.5)     #this way we can limit the axes in bar graph
plt.show()
```
