## Implementation of cat boost without encoding the dataset

```python
df1.drop(["Date"],axis=1,inplace=True)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
X=df1.drop(["Rented Bike Count"],axis=1)
Y=df1["Rented Bike Count"]
lr_model=LinearRegression()
x_train,x_test,y_train,y_test=train_test_split(X,Y,train_size=0.8)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from catboost import CatBoostRegressor
```

```python
cat_features =["Seasons","Functioning Day","Holiday"]
```

```python
one_hot_max_size = 5
ctb=CatBoostRegressor(iterations=100, depth=6, learning_rate=0.1,
                    verbose=False, cat_features=cat_features, one_hot_max_size=one_hot_max_size,loss_function='RMSE')
```

```python
ctb.fit(x_train,y_train)
y_predict_train = ctb.predict(x_train)
y_predict_test = ctb.predict(x_test)
print("Train R2:", r2_score(y_train, y_predict_train))
print("Test R2 :", r2_score(y_test, y_predict_test))
```
