```python
from abc import ABC, abstractmethod

import math
 
# 1. Abstraction

class AbstractCalculator(ABC):

    @abstractmethod

    def calculate(self):

        pass
 
# 2. Basic Calculator (Encapsulation + Inheritance)

class BasicCalculator(AbstractCalculator):

    def __init__(self):

        self.__result = 0  # Encapsulated private variable
 
    # Getter and setter (Encapsulation)

    def get_result(self):

        return self.__result
 
    def set_result(self, value):

        self.__result = value
 
    def calculate(self, a, b, operation):

        if operation == "add":

            self.__result = a + b

        elif operation == "subtract":

            self.__result = a - b

        elif operation == "multiply":

            self.__result = a * b

        elif operation == "divide":

            if b != 0:

                self.__result = a / b

            else:

                return "Error: Division by zero"
        else:
            return "Invalid basic operation"

        return self.__result
 
# 3. Scientific Calculator (Inheritance + Polymorphism)

class ScientificCalculator(BasicCalculator):

    # Overriding method (Polymorphism)

    def calculate(self, a, b=None, operation=None):

        if operation == "power":

            return math.pow(a, b)

        elif operation == "sqrt":

            return math.sqrt(a)

        elif operation == "log":

            return math.log(a)

        elif operation == "sin":

            return math.sin(math.radians(a))

        elif operation == "cos":

            return math.cos(math.radians(a))

        elif operation == "tan":

            return math.tan(math.radians(a))

        else:

            return super().calculate(a, b, operation)  # Reuse from BasicCalculator
 
# 4. Driver class

def calculator_menu():

    calc = ScientificCalculator()
 
    while True:

        print("\n===== Calculator Menu =====")

        print("1. Add")

        print("2. Subtract")

        print("3. Multiply")

        print("4. Divide")

        print("5. Power (x^y)")

        print("6. Square Root")

        print("7. Logarithm")

        print("8. Sine")

        print("9. Cosine")

        print("10. Tangent")

        print("0. Exit")
 
        choice = input("Choose an operation (0-10): ")
 
        if choice == "0":

            print("Exiting calculator...")

            break
 
        if choice in ["1", "2", "3", "4", "5"]:

            a = float(input("Enter first number: "))

            b = float(input("Enter second number: "))

        elif choice in ["6", "7", "8", "9", "10"]:

            a = float(input("Enter the number: "))

            b = None

        else:

            print("Invalid option!")

            continue
 
        operation_map = {

            "1": "add",

            "2": "subtract",

            "3": "multiply",

            "4": "divide",

            "5": "power",

            "6": "sqrt",

            "7": "log",

            "8": "sin",

            "9": "cos",

            "10": "tan"

        }
 
        operation = operation_map.get(choice)

        result = calc.calculate(a, b, operation)

        print(f"Result: {result}")
 
# Run the calculator

calculator_menu()
```

1. Create a Rectangle class

Attributes: length, width

Methods:

area() → returns area

perimeter() → returns perimeter

```python
class Rectangle:
    def __init__(self,height,width):
        self.height=height
        self.width=width

    def area(self):
        return self.height*self.width
    def perimeter(self):
        return 2* (self.height+self.width)
    
rect=Rectangle(10,20)
print(rect.area(),rect.perimeter())
```

3️⃣ Create a BankAccount class

Attributes: account_number, balance

Methods:

deposit(amount)

withdraw(amount)

display_balance()

```python
class BankAccount:
    def __init__(self,account_number,balance):
        self.account_number=account_number
        self.balance=balance

    def deposit(self,amount):
        self.amount=amount
        self.balance+=self.amount

    def withdraw(self,amount):
        self.amount=amount
        self.balance-=self.amount

    def display_balance(self):
        return self.balance

c1=BankAccount(100,5000)
c1.deposit(2000)

c2=BankAccount(101,100000)
c2.withdraw(1000)

print("Balance is",c1.display_balance())
print("Balance is",c2.display_balance())
```

Use staticmethods so you can call methods without creating an object.

```python
class Calculator :
    @staticmethod
    def add(a,b):
        return a+b
    @staticmethod
    def subtract(a,b):
        return a-b
    @staticmethod
    def multiply(a,b):
        return a*b
    @staticmethod
    def divide(a,b):
        return a/b
    
obj1=Calculator()
print("Sum of two Numbers",obj1.add(3,5))
print("Difference of two Numbers",obj1.subtract(321123123,123122314412))
print("Multiplication of two Numbers",obj1.multiply(342,432))
print("Division of two Numbers",obj1.divide(12344,34))
```

7️⃣ Create a class Circle with encapsulation

Private attribute: __radius

Method: get_radius() and set_radius() to access/modify it.

Add a method to calculate area of circle.

```python
import math
class Circle:
    def __init__(self,radius):
        self.__radius=radius
        
    def get_radius(self):
        return self.__radius
    def set_radius(self,new_radius):
        self.__radius=new_radius

    def area(self):
        area= math.pi*(self.__radius*self.__radius)
        return area
    
c1=Circle(5)
print("Radius of Circle is",c1.get_radius())
print("Area of circle ",c1.area())
(c1.set_radius(6))
print("Area of circle", c1.area(),"with radius of ",c1.get_radius())
```

8️⃣ Create an abstract class Shape

Abstract method: area()

Subclasses: Square and Triangle must implement area().

```python
from abc import ABC, abstractmethod 

class Shape(ABC):
    def __init__(self):
        pass

    @abstractmethod
    def area(self):
        return "You are a Fool!!!"


class Triangle(Shape):
    def __init__(self,height,base):
        self.height=height
        self.base=base

    def area(self):
        return 0.5*self.base*self.height
    
class Square(Shape):
    def __init__(self,side):
        self.side=side
    
    def area(self):
        return self.side**2
    

t1=Triangle(4,5)
print("Area of triangle is ",t1.area())

s1=Square(6)
print("Area of Square is ",s1.area())
```

9️⃣ Create classes Teacher and Researcher

Both have a method show_role()

Create a subclass Professor(Teacher, Researcher) and observe how MRO (Method Resolution Order) works.

```python
class Teacher:
    def __init__(self,role):
        self.role=role
    def show_role(self):
        return f"{self.role} This is Teacher class"
    
class Researcher:
    def __init__(self,role):
        self.role=role

    def show_role(self):
        return f"{self.role} This is Researcher class"
    
class Professor(Teacher,Researcher):
    def __init__(self, role):
        super().__init__(role)
        

p1=Professor("Educator")
print(p1.show_role()) # MRO precedence is based on L->R of classes written while inheritance
```

🔟 Create a class Library

Uses composition to include a list of Book objects.

Each Book has attributes: title, author, price.

Add methods in Library to:

Add a book 

Remove a book

Display all books

```python
class Book:
    def __init__(self,title,author,price):
        self.title=title
        self.author=author
        self.price=price

    def show_details(self):
        return self.title,self.author,self.price
    
class Library():
    def  __init__(self):
        self.books=[]

    def add_book(self,book):
        self.books.append(book)
        return f"Added {book.title} "
    def remove_book(self,book):
        self.books.remove(book)
        return f"Removed {book.title}"
    def show_books(self):
        for book in self.books:
            print( f"{book.title} by {book.author} is of Price {book.price}")
        

b1=Book("OOPS","Lokesh Logesh JIII",-2000)
b2=Book("Maha Java","Amul Mukul Masti",1000)

l=Library()
l.add_book(b1)
l.add_book(b2)
l.show_books()
```
