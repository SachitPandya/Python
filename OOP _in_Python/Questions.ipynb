```python
from abc import ABC, abstractmethod

import math
 
# 1. Abstraction

class AbstractCalculator(ABC):

    @abstractmethod

    def calculate(self):

        pass
 
# 2. Basic Calculator (Encapsulation + Inheritance)

class BasicCalculator(AbstractCalculator):

    def __init__(self):

        self.__result = 0  # Encapsulated private variable
 
    # Getter and setter (Encapsulation)

    def get_result(self):

        return self.__result
 
    def set_result(self, value):

        self.__result = value
 
    def calculate(self, a, b, operation):

        if operation == "add":

            self.__result = a + b

        elif operation == "subtract":

            self.__result = a - b

        elif operation == "multiply":

            self.__result = a * b

        elif operation == "divide":

            if b != 0:

                self.__result = a / b

            else:

                return "Error: Division by zero"
        else:
            return "Invalid basic operation"

        return self.__result
 
# 3. Scientific Calculator (Inheritance + Polymorphism)

class ScientificCalculator(BasicCalculator):

    # Overriding method (Polymorphism)

    def calculate(self, a, b=None, operation=None):

        if operation == "power":

            return math.pow(a, b)

        elif operation == "sqrt":

            return math.sqrt(a)

        elif operation == "log":

            return math.log(a)

        elif operation == "sin":

            return math.sin(math.radians(a))

        elif operation == "cos":

            return math.cos(math.radians(a))

        elif operation == "tan":

            return math.tan(math.radians(a))

        else:

            return super().calculate(a, b, operation)  # Reuse from BasicCalculator
 
# 4. Driver class

def calculator_menu():

    calc = ScientificCalculator()
 
    while True:

        print("\n===== Calculator Menu =====")

        print("1. Add")

        print("2. Subtract")

        print("3. Multiply")

        print("4. Divide")

        print("5. Power (x^y)")

        print("6. Square Root")

        print("7. Logarithm")

        print("8. Sine")

        print("9. Cosine")

        print("10. Tangent")

        print("0. Exit")
 
        choice = input("Choose an operation (0-10): ")
 
        if choice == "0":

            print("Exiting calculator...")

            break
 
        if choice in ["1", "2", "3", "4", "5"]:

            a = float(input("Enter first number: "))

            b = float(input("Enter second number: "))

        elif choice in ["6", "7", "8", "9", "10"]:

            a = float(input("Enter the number: "))

            b = None

        else:

            print("Invalid option!")

            continue
 
        operation_map = {

            "1": "add",

            "2": "subtract",

            "3": "multiply",

            "4": "divide",

            "5": "power",

            "6": "sqrt",

            "7": "log",

            "8": "sin",

            "9": "cos",

            "10": "tan"

        }
 
        operation = operation_map.get(choice)

        result = calc.calculate(a, b, operation)

        print(f"Result: {result}")
 
# Run the calculator

calculator_menu()
```

```python
from abc import ABC , abstractmethod
class PaymentMethod(ABC):
    def __init__(self):
        self.amount = 0
       
    @abstractmethod
    def processPayment(self):
        pass
   
class Discountable(PaymentMethod):
    def applyDiscount(self,amount,discount):
        self.discount = discount
        self.amount=amount
        self.amount -= self.amount*(discount/100)
        return self.amount
   
class CreditCard(Discountable):
 
    def processPayment(self,amount,discount):
        super().applyDiscount(amount,discount)
        return "the payment is done with CreditCard , the discounted amount is" , self.amount
 
class PayPal(Discountable):
 
    def processPayment(self,amount,discount):
        self.amount = amount
        super().applyDiscount(amount,discount)
        return "the payment is done with PayPal , the discounted amount is" , self.amount
 
a1 = CreditCard()
print(a1.processPayment(1500,30))
a2=PayPal()
print(a2.processPayment(1600,20))
```
