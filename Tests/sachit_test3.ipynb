1. Create a function that takes a sentence as input, counts the number of vowels, and returns the count.

```python
# this function is used to get the count of vowels in the 
def count_vowel(input_string):                  
    vowels=['a','e','i','o','u']
    counter=0

    for index in input_string:
        for vowel in vowels:
            if index==vowel:
                counter+=1
    
    return counter

given_string=input("enter string")
number_of_vowels=count_vowel(given_string)
print("There are",number_of_vowels,"vowels in the sentence -> ",given_string)
```

2. Implement a function that calculates the square root of each element in a list and returns a new list without inbuilt functions and use recursive function.

```python
def square_root(num):
    lst=[]
    for j in num:
        for i in range(1,j):
            if j%i==0 and i*i==j:
                lst+=[i]
                break
    return lst

input_list=[4,9,16,25]
sqrt_list=square_root(input_list)
print(sqrt_list)    
```

3. Create a generator function that yields the powers of 2 up to a certain limit.

```python
def pow_of_two(num):
    for pow in range(num+1):
        yield 2**pow

upper_limit=int(input("enter the number upto which you want power of 2"))
powers=pow_of_two(upper_limit)
for i in range(upper_limit+1):
    print(next(powers),end=" ")
```

5. Create a generator function that yields all the permutations of a given list

```python
def fact(n):
    factorial=0
    if n==1:
        return 1
    else:
        for i in range(n):
            factorial+=fact(n-1)
    return factorial

def permutation(given_list):
    length=len(given_list)
    return fact(length)

input_list=[1,2,3,4,5,6]
x=permutation(input_list)
print(x)
```

```python
import itertools

input_list=[1,2,3,4,5,6]

x=list(itertools.permutations(input_list))
print(x,len(x))
```

6. Implement a generator function to flatten a nested list.

```python
nested_list=[[1,2],[2,4],[5,6]]

def flatten_list(given_list):
    for i in given_list:
        if type(i)==list:
            for j in i:
                yield j
        else:
            yield i

elements=flatten_list(nested_list)
try:
    while True:
        print(next(elements),end=" ")
except StopIteration:
    pass
```
