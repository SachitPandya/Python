4. Merge the relevant data on EmployeeID and DepartmentID columns to create a new column 'Total_Salary' by summing the Salary for each department.

```python
Department_data=stock_purchase_data.groupby(["DepartmentID"])     #grouped the data on the beasis of DepartmentID to calculate the total department salary
employee_salary=Department_data[['Salary']]                       #sliced a particular column from this grouping to get the total sum
Department_total_salary=employee_salary.sum()                   #performed the task to get total sum

stock_purchase_data=pd.merge(stock_purchase_data,Department_total_salary,on="DepartmentID",how='inner')     
#inner joined the total salary department wise to get the total salary of department in front department employee

#renaming the columns as during inner join names of the salary column got changed
stock_purchase_data.rename(columns={"Salary_x":"Salary","Salary_y":"Total_Salary"},inplace=True)

#printing the DataFrame
stock_purchase_data
```

5. Identify the top 3 sensors with the highest average value, excluding sensors with missing values for more than 10% of the timestamps.
