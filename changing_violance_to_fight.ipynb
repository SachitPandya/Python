```python
import os
import shutil
import pandas as pd

# Input/output directories
input_folder = r"C:\Users\BAPS\Desktop\fighting_dataset\t_dataset\valid"
output_folder = r"C:\Users\BAPS\Desktop\fighting_dataset\t_dataset\train_violance"
csv_path = r"C:\Users\BAPS\Desktop\fighting_dataset\t_dataset\train\_classes.csv" # path to your CSV file
filtered_csv_path = r"C:\Users\BAPS\Desktop\fighting_dataset\t_dataset\train\_classes.csv"  # new CSV after filtering

# Keywords
violence_keywords = ["violence"]
non_violence_keywords = ["non-violence", "non_violence", "nonviolence", "no-violence", "no_violence"]

# Step 1: Ensure output folder exists
os.makedirs(output_folder, exist_ok=True)

# Step 2: Filter files
copied_count = 0
deleted_count = 0

for filename in os.listdir(input_folder):
    file_path = os.path.join(input_folder, filename)

    if not os.path.isfile(file_path):
        continue

    name_lower = filename.lower()

    if any(non in name_lower for non in non_violence_keywords):
        os.remove(file_path)
        deleted_count += 1
        print(f"Deleted: {filename}")
        continue

    if any(violence in name_lower for violence in violence_keywords):
        shutil.copy(file_path, output_folder)
        copied_count += 1

# Step 3: Filter CSV
df = pd.read_csv(csv_path)

# Convert the relevant column (assume it's "filename" or "label") to lowercase
df["filename_lower"] = df["filename"].str.lower()

# Filter out rows with any non-violence keyword
for keyword in non_violence_keywords:
    df = df[~df["filename_lower"].str.contains(keyword, na=False)]

# Drop helper column
df = df.drop(columns=["filename_lower"])

# Save cleaned CSV
df.to_csv(filtered_csv_path, index=False)

print(f"\nTotal violence images copied: {copied_count}")
print(f"Total non-violence images deleted: {deleted_count}")
print(f"Filtered CSV saved to: {filtered_csv_path}")
```

```python
import os
import shutil
import xml.etree.ElementTree as ET
 
 
class_name = "fig"
source_dir = r"C:\Users\BAPS\Desktop\extracting_frames\frames"   # Folder with new raw files
target_dir = r"C:\Users\BAPS\Desktop\extracting_frames\frames11"  # Final folder
 
# Valid image types
valid_exts = [".jpg", ".jpeg", ".png"]
 
 
 
def get_max_existing_count(directory, class_name):
    max_count = 0
    for root, _, files in os.walk(directory):
        for file in files:
            if file.startswith(f"{class_name}_") and file.lower().endswith(tuple(valid_exts)):
                try:
                    num = int(file.split("_")[1].split(".")[0])
                    max_count = max(max_count, num)
                except:
                    continue
    return max_count + 1
 
def update_xml_filename(xml_path, new_filename):
    try:
        tree = ET.parse(xml_path)
        root = tree.getroot()
        filename_tag = root.find("filename")
        if filename_tag is not None:
            filename_tag.text = new_filename
        tree.write(xml_path)
    except Exception as e:
        print(f"Error updating XML {xml_path}: {e}")
 
def move_source_to_target(source_dir, target_dir):
    os.makedirs(target_dir, exist_ok=True)
    all_files = os.listdir(source_dir)
    moved = 0
 
    for file in all_files:
        if file.lower().endswith(tuple(valid_exts)) or file.lower().endswith(".xml"):
            src_path = os.path.join(source_dir, file)
            dst_path = os.path.join(target_dir, file)
            shutil.move(src_path, dst_path)
            moved += 1
 
    print(f"Moved {moved} files from source to target.\n")
 
def rename_files(folder_path, class_name):
    count = get_max_existing_count(folder_path, class_name)
    renamed = 0
 
    for root, _, files in os.walk(folder_path):
        jpg_files = {f.split('.')[0]: f for f in files if f.lower().endswith(tuple(valid_exts))}
        xml_files = {f.split('.')[0]: f for f in files if f.lower().endswith(".xml")}
        common_keys = jpg_files.keys() & xml_files.keys()
 
        for key in sorted(common_keys):
            img_file = jpg_files[key]
            xml_file = xml_files[key]
 
            # Skip already renamed files
            if img_file.startswith(f"{class_name}_") and xml_file.startswith(f"{class_name}_"):
                continue
 
            img_ext = os.path.splitext(img_file)[1].lower()
            old_img_path = os.path.join(root, img_file)
            old_xml_path = os.path.join(root, xml_file)
 
            new_name = f"{class_name}_{count}"
            new_img_file = new_name + img_ext
            new_xml_file = new_name + ".xml"
 
            new_img_path = os.path.join(root, new_img_file)
            new_xml_path = os.path.join(root, new_xml_file)
 
            try:
                os.rename(old_img_path, new_img_path)
                os.rename(old_xml_path, new_xml_path)
                update_xml_filename(new_xml_path, new_img_file)
                print(f"Renamed: {img_file} + {xml_file} -> {new_img_file} + {new_xml_file}")
                count += 1
                renamed += 1
            except Exception as e:
                print(f"Error renaming {img_file}: {e}")
 
    print(f"Total renamed pairs: {renamed}")


move_source_to_target(source_dir, target_dir)
rename_files(target_dir, class_name)
```

### Resolution 

```python
import os
import shutil
import xml.etree.ElementTree as ET

def filter_by_resolution(
    source_dir,
    output_dir,
    min_width=360,
    min_height=360
):
    os.makedirs(output_dir, exist_ok=True)
    removed_files = []

    all_files = os.listdir(source_dir)
    jpg_files = {f.split('.')[0]: f for f in all_files if f.endswith('.jpg')}
    xml_files = {f.split('.')[0]: f for f in all_files if f.endswith('.xml')}
    common_keys = jpg_files.keys() & xml_files.keys()

    for key in common_keys:
        xml_path = os.path.join(source_dir, xml_files[key])
        jpg_path = os.path.join(source_dir, jpg_files[key])

        try:
            tree = ET.parse(xml_path)
            root = tree.getroot()
            size = root.find("size")
            if size is not None:
                width = int(size.find("width").text)
                height = int(size.find("height").text)
                if width >= min_width and height >= min_height:
                    #  Copy image and annotation
                    shutil.copy(xml_path, output_dir)
                    shutil.copy(jpg_path, output_dir)
                else:
                    removed_files.append(key)
            else:
                removed_files.append(key)
        except Exception as e:
            print(f" Error processing {xml_path}: {e}")
            removed_files.append(key)

    # Save removed list
    removed_list_path = os.path.join(output_dir, "removed_due_to_resolution.txt")
    with open(removed_list_path, "w") as f:
        for key in removed_files:
            f.write(f"{key}\n")

    print(f" Saved {len(common_keys) - len(removed_files)} valid files.")
    print(f" Removed {len(removed_files)} due to resolution < {min_width}x{min_height}.")
    print(f" Skipped filenames saved to: {removed_list_path}")

#  Example usage
source =  r"C:\Users\BAPS\Desktop\fighting_frames\New folder\image_directory\image_totals1"
destination =  r"C:\Users\BAPS\Desktop\fighting_frames\New folder\image_directory\image_total" 
filter_by_resolution(source, destination, min_width=360, min_height=360)
```

```python
import os
from PIL import Image
import imagehash
import shutil

def find_and_store_excess_duplicates(folder_path, backup_folder, keep_count=1):
    image_hashes = {}  # hash: list of file paths
    moved_files = []   # (hash, moved_file_path)

    if not os.path.exists(backup_folder):
        os.makedirs(backup_folder)

    # Phase 1: Detect and move excess duplicates
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif')):
                file_path = os.path.join(root, file)
                try:
                    img = Image.open(file_path)
                    hash_val = str(imagehash.average_hash(img))  # Convert hash to string for logging and key

                    # Store file paths under this hash
                    if hash_val in image_hashes:
                        image_hashes[hash_val].append(file_path)

                        # If count exceeds keep_count, move the extra
                        if len(image_hashes[hash_val]) > keep_count:
                            backup_path = os.path.join(backup_folder, file)
                            shutil.move(file_path, backup_path)
                            moved_files.append((hash_val, file_path))
                            print(f"Moved extra duplicate to backup: {file_path}")
                    else:
                        image_hashes[hash_val] = [file_path]

                except Exception as e:
                    print(f"Error processing {file_path}: {e}")

    # Phase 2: Logging removed duplicates
    if moved_files:
        print("\n=== Removed Duplicate Files (Exceeding 4 per Image) ===")
        for hash_val in set([item[0] for item in moved_files]):
            related_files = [f for h, f in moved_files if h == hash_val]
            print(f"\nHash: {hash_val}")
            for fpath in related_files:
                print(f"  Removed: {fpath}")
    else:
        print("No extra duplicates found.")

    # Phase 3: Delete XML files without corresponding JPG
    print("\n=== Checking for orphaned XML files ===")
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith('.xml'):
                xml_path = os.path.join(root, file)
                jpg_name = os.path.splitext(file)[0] + '.jpg'
                jpg_path = os.path.join(root, jpg_name)

                if not os.path.exists(jpg_path):
                    try:
                        os.remove(xml_path)
                        print(f"Deleted orphaned XML: {xml_path}")
                    except Exception as e:
                        print(f"Error deleting {xml_path}: {e}")

    return moved_files

# Example usage
source_folder = r"C:\Users\BAPS\Desktop\extracting_frames\frames"
backup_folder = r"C:\Users\BAPS\Desktop\extracting_frames\duplicates_backup"

removed = find_and_store_excess_duplicates(source_folder, backup_folder)

print(f"\nTotal extra duplicates removed: {len(removed)}")
```

```python
import os
import shutil
import xml.etree.ElementTree as ET
from PIL import Image
import imagehash
 
def filter_images_by_resolution(source_dir, temp_dir, min_width=360, min_height=360):
    os.makedirs(temp_dir, exist_ok=True)
    removed_dir = os.path.join(temp_dir, "removed_files")
    os.makedirs(removed_dir, exist_ok=True)
 
    all_files = os.listdir(source_dir)
    jpg_files = {os.path.splitext(f)[0].lower(): f for f in all_files if f.lower().endswith('.jpg')}
    xml_files = {os.path.splitext(f)[0].lower(): f for f in all_files if f.lower().endswith('.xml')}
    common_keys = jpg_files.keys() & xml_files.keys()
 
    for key in common_keys:
        jpg_path = os.path.join(source_dir, jpg_files[key])
        xml_path = os.path.join(source_dir, xml_files[key])
        try:
            tree = ET.parse(xml_path)
            root = tree.getroot()
            size = root.find("size")
            if size is not None:
                width = int(size.find("width").text)
                height = int(size.find("height").text)
                if width >= min_width and height >= min_height:
                    shutil.copy(jpg_path, temp_dir)
                    shutil.copy(xml_path, temp_dir)
                else:
                    shutil.copy(jpg_path, removed_dir)
                    shutil.copy(xml_path, removed_dir)
            else:
                shutil.copy(jpg_path, removed_dir)
                shutil.copy(xml_path, removed_dir)
        except:
            shutil.copy(jpg_path, removed_dir)
            shutil.copy(xml_path, removed_dir)
 
    # Handle images without XML
    for key, jpg_file in jpg_files.items():
        if key not in common_keys:
            jpg_path = os.path.join(source_dir, jpg_file)
            try:
                img = Image.open(jpg_path)
                width, height = img.size
                if width >= min_width and height >= min_height:
                    shutil.copy(jpg_path, temp_dir)
                else:
                    shutil.copy(jpg_path, removed_dir)
            except:
                shutil.copy(jpg_path, removed_dir)
 
def remove_excess_duplicates(folder_path, backup_folder, keep_count=4):
    os.makedirs(backup_folder, exist_ok=True)
    image_hashes = {}
    for root, _, files in os.walk(folder_path):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png', '.bmp', '.gif')):
                file_path = os.path.join(root, file)
                try:
                    img = Image.open(file_path)
                    hash_val = str(imagehash.average_hash(img))
                    image_hashes.setdefault(hash_val, []).append(file_path)
                except:
                    continue
 
    for hash_val, paths in image_hashes.items():
        if len(paths) > keep_count:
            for extra_path in paths[keep_count:]:
                shutil.move(extra_path, os.path.join(backup_folder, os.path.basename(extra_path)))
                xml_path = os.path.splitext(extra_path)[0] + ".xml"
                if os.path.exists(xml_path):
                    shutil.move(xml_path, os.path.join(backup_folder, os.path.basename(xml_path)))
 
# === Usage ===
source_dir = r"C:\Users\BAPS\Desktop\fighting_dataset\t_dataset\test"
temp_dir = r"C:\Users\BAPS\Desktop\fighting_dataset\t_dataset\high_resolution"
backup_dir = r"C:\Users\BAPS\Desktop\fighting_dataset\t_dataset\duplicates"
 
filter_images_by_resolution(source_dir, temp_dir)
remove_excess_duplicates(temp_dir, backup_dir)
```
